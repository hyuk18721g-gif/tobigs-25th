## A1. 답안 작성  
A/B TEST란, 두 가지의 대안을 실제 사용자에게 무작위로 노출해 어떤 안이 더 좋은 성과를 내는지 통계적으로 검증하는 실험 방법이다.
A/B TEST 진행 시 대표적으로 유의할 점은 4가지이다.
1. 명확한 가설과 단일 변경: 하나의 요소만 타겟으로 삼고, 명확히 세운 가설을 검정해야 정확한 효과를 확인할 수 있다.
2. 충분한 수의 표본을 무작위 배정: 사용자 특성(연령, 지역, 기기 등)이 A/B에 골고루 분산돼야 한다.
3. 충분한 실험 기간 : 사용자에게 나타날 수 있는 초두효과(처음 접한 정보나 디자인에 더 큰 영향을 받는 현상)를 견제하기 위해, 초기 반응이 아니라 안정화된 행동을 관찰할 수 있는 충분한 기간이 필요하다.
4. 외부 요인 통제 : 실험 중 이벤트가 발생해 영향을 받는 걸 조심하고, 가능하면 동시에 실험하는 게 좋다.

## Q2-2. 실험에 필요한 최소 test users 수를 구해봅시다.
유의수준 0.05, 검정력 80% 하에 수행해주세요.  


아래 공식 참고해서, 최소 test users 수를 구하는 코드 작성해주세요.

추가적으로 n은 산출하는데에 가정한 p1, p2도 같이 작성해주세요.

$$n = \frac{(Z_{\alpha/2} + Z_{\beta})^2 \cdot [p_1(1-p_1) + p_2(1-p_2)]}{(p_1 - p_2)^2}$$

## A2-2. 아래 질문에 대한 답과 코드를 채워주세요.

p1, p2는 각각 무엇을 의미하는 확률이고, 어떤 값으로 가정하셨나요?   
또한 위 공식을 보고 n을 구하기 위한 코드를 작성해주세요. 단, 라이브러리는 z값을 구하는 용도의 것만 사용해주세요.


### 답변:
p1 ​= A안 전체 CVR = 0.30⋅0.40+0.70⋅0.25 = 0.12+0.175 = 0.295

p2 = B안 전체 CVR = 0.50⋅0.40+0.50⋅0.25 = 0.20+0.125 = 0.325

```python
# A2-2 코드작성. 나온 n은 아래 Minimum Sample Size에 기재.

from scipy.stats import norm
import math

# =========================
# 1) 실험 조건
# =========================
alpha = 0.05      # 유의수준
power = 0.80      # 검정력(=1-β)

Z_alpha_over_2 = norm.ppf(1 - alpha/2)   # Z_{α/2} = Z_{0.975}
Z_beta = norm.ppf(power)                 # Z_{β} = Z_{0.8}

# =========================
# 2) p1, p2 가정 (문제 맥락 기반)
# =========================
p_tab = 0.40          # 탭 이용자 CVR
p_no_tab = 0.25       # 탭 미이용자 CVR

tab_rate_A = 0.30     # 현재 탭 이용률 (A안)
tab_rate_B = 0.50     # 개선 후 탭 이용률 가정 (B안)

p1 = tab_rate_A * p_tab + (1 - tab_rate_A) * p_no_tab
p2 = tab_rate_B * p_tab + (1 - tab_rate_B) * p_no_tab

# =========================
# 3) 표본수 계산 (그룹당 n)
# =========================
numerator = (Z_alpha_over_2 + Z_beta)**2 * (p1*(1-p1) + p2*(1-p2))
denominator = (p1 - p2)**2
n_per_group = numerator / denominator

print(f"Z_(alpha/2) = {Z_alpha_over_2:.4f}, Z_beta = {Z_beta:.4f}")
print(f"p1 = {p1:.4f}, p2 = {p2:.4f}, absolute lift = {p2-p1:.4f}")
print(f"Required n per group = {math.ceil(n_per_group)}")
print(f"Total users (A+B) = {math.ceil(2*n_per_group)}")
```

 Minimum Sample Size = 3727

귀무가설 ($H_0$): 메인 홈 검색 모듈의 위치(하단 vs 최상단)와 상세 페이지 진입 여부 사이에는 상관관계가 없다. 즉, 두 그룹의 상세 페이지 진입률($CVR$)은 동일하다.$$H_0: p_A = p_B$$대립가설 ($H_1$): 메인 홈 검색 모듈을 최상단에 노출하는 것은 상세 페이지 진입 여부에 영향을 미친다. 즉, 실험군(B)의 상세 페이지 진입률이 대조군(A)보다 높을 것이다.$$H_1: p_A < p_B$$

## A3-1.  검정 이름과 이유를 설명해주세요.

답변: 위 가설에 필요한 검정은 카이제곱 검정이고,  

이유 : 본 실험의 반응 변수는 상세 페이지 진입 여부로 이진형 변수이며, 관심 대상은 평균이 아닌 두 집단 간 진입률(CVR)의 차이다. 따라서 두 범주형 변수의 독립성을 검정하는 카이제곱 검정이 더 적절하다.

<br>

```python
'''''''''''''''''''''''''''''''''''''''
제가 세운 A/B TEST 설계 실험 코드입니다.
'''''''''''''''''''''''''''''''''''''''

import pandas as pd
from scipy.stats import chi2_contingency

# 1. 관측 데이터 입력 (실제 실험 결과 수치로 변경하세요)
# [상세진입 성공 수, 상세진입 실패 수]
control_results = [11500, 34500]  # 대조군(A)
variant_results = [12190, 33810]  # 실험군(B)

# 2. 분할표(Contingency Table) 생성
data = [control_results, variant_results]

# 3. 카이제곱 검정 수행
chi2, p_value, dof, expected = chi2_contingency(data)

# 4. 결과 출력
print(f"카이제곱 통계량: {chi2:.4f}")
print(f"P-value: {p_value:.10f}")

alpha = 0.05
if p_value < alpha:
    print(f"결과: P-value가 {alpha}보다 작으므로 귀무가설을 기각합니다.")
    print("=> 검색바 최상단 노출이 상세 페이지 진입률에 유의미한 변화를 주었습니다.")
else:
    print(f"결과: P-value가 {alpha}보다 크므로 귀무가설을 기각할 수 없습니다.")
    print("=> 검색바 위치 변경의 효과가 통계적으로 증명되지 않았습니다.")
```

## Q3-4. 위에서 답한 것을 토대로, 검정을 수행해주세요. (t test와 카이제곱 검정 중)


## A3-4 아래 셀에 코드 작성해주세요. 그리고!!!
그리고 나온 값을 통해 수행하신 a/b test 결과를 설정하신 가설에 맞게 해석해주세요.

```python
### A3-4. 코드작성

import numpy as np
from scipy.stats import chi2_contingency

# =========================
# 1) 관측 데이터 입력
# =========================
# [상세진입 성공 수, 상세진입 실패 수]
control_results = [11500, 34500]  # 대조군(A)
variant_results = [12190, 33810]  # 실험군(B)

# 2x2 분할표
table = np.array([control_results, variant_results])

# =========================
# 2) 비율 계산 (방향성 확인용)
# =========================
a_success, a_fail = control_results
b_success, b_fail = variant_results

nA = a_success + a_fail
nB = b_success + b_fail

pA_hat = a_success / nA
pB_hat = b_success / nB
diff = pB_hat - pA_hat

print(f"pA_hat = {pA_hat:.6f}, pB_hat = {pB_hat:.6f}, (pB_hat - pA_hat) = {diff:.6f}")

# =========================
# 3) 카이제곱 검정 (양측 p-value)
# =========================
chi2, p_two, dof, expected = chi2_contingency(table, correction=False)

# =========================
# 4) 단측 가설(H1: pA < pB)로 p-value 변환
# =========================
# 방향이 H1과 같은지 확인 후 단측 p-value 구성
if pB_hat > pA_hat:
    p_one = p_two / 2
else:
    p_one = 1 - (p_two / 2)

print(f"Chi-square = {chi2:.4f}, p(two-sided) = {p_two:.10f}, p(one-sided, H1: pA<pB) = {p_one:.10f}")

# =========================
# 5) 결론
# =========================
alpha = 0.05
if p_one < alpha:
    print(f"결과: p(one-sided) < {alpha} 이므로 귀무가설(H0: pA=pB)을 기각합니다.")
    print("=> 검색바 상단 노출(B)이 상세 페이지 진입률(CVR)을 유의미하게 증가시킵니다. (H1: pA < pB 지지)")
else:
    print(f"결과: p(one-sided) ≥ {alpha} 이므로 귀무가설(H0)을 기각할 수 없습니다.")
    print("=> 상단 노출(B)의 CVR 증가 효과가 통계적으로 충분히 입증되지 않았습니다.")
```

**결과해석**

답변:

1. 관측 결과, 대조군(A)의 상세 페이지 진입률(CVR): 25.0%, 실험군(B)의 상세 페이지 진입률(CVR): 26.5%로 실험군(B)이 대조군(A) 대비 더 높은 상세 페이지 진입률을 보였다.
2. 통계적 검정 결과, 카이제곱 통계량: χ² = 27.0668, 양측 p-value: 1.965 × 10⁻⁷, 단측 p-value (H₁: pA < pB): 9.83 × 10⁻⁸으로 관측된 CVR 차이가 H0가 참이라는 가정 하에 발생했을 가능성은 극히 낮다.

=> 따라서 검색바를 상단에 노출한 실험군의 상세 페이지 진입률은 대조군보다 유의미하게 높았으며(p < 0.001), 이에 따라 검색바 상단 노출은 사용자 전환을 개선하는 효과적인 UI 변경으로 판단된다.
